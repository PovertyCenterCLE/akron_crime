**A few notes about the data.** The data was initially filtered by the top level NIBRS code, such that crime reports with multiple NIBRS codes would select the "lowest code" -- perhaps referring to the most severe crime, however there are a few Part 2 crimes that are lower on this list than Part 1 crimes, where we define Part 1 crimes as potentially "worse" than Part 2. This difference is just in how we catagorize crimes may still need to be standardized. 

```python
import pandas as pd
```

```python
df_cv = pd.read_csv('cv_crime.csv', index_col = [0], parse_dates = [4])
df_ed = pd.read_csv('ed_crime.csv', index_col = [0], parse_dates = [4])
df_filtered = pd.read_csv('summit_crime_filtered_ungeocoded.csv', index_col = [0])
```

```python
cv = pd.merge(df_filtered, df_cv[['address','x','y']],how='inner',on=['address'])
```

```python
ed = pd.merge(df_filtered, df_ed[['address','x','y']],how='inner',on=['address'])
```

```python
nibrs= ['09A','09B','09C','100','11A','11B','11C','11D','120','13A','13B','13C','200','210','220','23A','23B','23C','23D','23E','23F', \
        '23G','23H','240','250','26A','26B','26C','26D','26E','270','280','290','35A','35B','36A','36B','370','39A','39B','39C','39D', \
        '40A','40B','510','520','90A','90B','90C','90D','90E','90F','90G','90H','90I','90J','90Z']

part1 = ['09A','09B','09C','11A','11B','11D','120','13A','200','210','220','23A','23B','23C','23D','23E','23F','23G','23H','240','36B']
part2 = list(set(nibrs)-set(part1))
```

```python
cv['year'] = cv.datetime_occurred.apply(lambda x: int(x[0:4]))
ed['year'] = ed.datetime_occurred.apply(lambda x: int(x[0:4]))
```

```python
cv = cv[(cv.year>1994) & (cv.year<2015)]
ed = ed[(ed.year>1994) & (ed.year<2015)]
```

```python
def catagorize(x):
    if x in part1:
        return 1
    else:
        return 0
cv['part1'] = cv['nibrs_code'].apply(catagorize)
ed['part1'] = ed['nibrs_code'].apply(catagorize)
```

```python
cv_part1 = cv.loc[cv.part1==1, ['nibrs_code', 'year']].groupby('year').count()
cv_part2 = cv.loc[cv.part1==0, ['nibrs_code', 'year']].groupby('year').count()
```

```python
ed_part1 = ed.loc[ed.part1==1, ['nibrs_code', 'year']].groupby('year').count()
ed_part2 = ed.loc[ed.part1==0, ['nibrs_code', 'year']].groupby('year').count()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('fivethirtyeight')
```

```python
ax = plt.subplot(1,1,1)
plt.plot(cv_part1.index, cv_part1.values,color = (0.5,0,1,0.7))
plt.plot(cv_part2.index, cv_part2.values,'--',color = (0.5,0,1,0.7))
plt.plot(ed_part1.index, ed_part1.values, color =  (0.2,.7,0.1,0.7))
plt.plot(ed_part2.index, ed_part2.values,'--', color = (0.2,.7,0.1,0.7))

plt.title('Crimes in Edgewood and Cascade Village by year', y=1.1, x = .65)
plt.legend(['Part 1 crimes, Cascade Village','Part 2 crimes, Cascade Village','Part 1 crimes, Edgewood','Part 2 crimes, Edgewood'], loc = (1.01, 0.2))

plt.xlabel('Year')
plt.ylabel('Count')

plt.xlim([1994,2015])
plt.ylim([-50,850])

plt.annotate("Note: Part 1 crimes are crimes with NIBRS code in 09A, 09B, 09C, 11A, 11B, 11D,\n         120, 13A, 200, 210, 220, 23A, 23B, 23C, 23D, 23E, 23F, 23G, 23H, 240, 36B", \
             (0,0),(0,-60), xycoords='axes fraction', textcoords='offset points', va='top')
```
